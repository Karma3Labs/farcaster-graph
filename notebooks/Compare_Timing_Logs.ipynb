```python
def read_timings(log_file):
    df = pd.read_csv(log_file, header=None, sep='\\s+')
    display(df.head())
    df.drop(df.columns[[0,1,2,3,4,5,7,9]], axis=1, inplace=True)
    display(df.head())
    df.rename(columns={6:'channel', 8:'timing'}, inplace=True)
    display(df.head())
    return df
```

```python
bad_timings = read_timings('/tmp/ch_rank_bad_timings.log')
```

```python
good_timings = read_timings('/tmp/ch_rank_good_timings.log')
```

```python
merged = good_timings.merge(bad_timings, on='channel', how='outer', suffixes=('_g', '_b'))
```
